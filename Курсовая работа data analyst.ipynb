```python
#2.4
progresss = main[main['is_homework']][['start_date', 'finish_date', 'title', 'progress_id']]
progresss['finish_date'] = pd.to_datetime(progresss['finish_date']).dt.date
progresss['finish_date'] = pd.to_datetime(progresss['finish_date'])
progresss['finish_date'] = progresss['finish_date'].apply(lambda dt: dt.replace(day=1))
progresss['start_date'] = pd.to_datetime(progresss['start_date']).dt.date
progresss['start_date'] = pd.to_datetime(progresss['start_date'])
progresss['start_date'] = progresss['start_date'].apply(lambda dt: dt.replace(day=1))
title=['Excel Базовый','Java-разработчик','Java-разработчик c нуля','JavaScript с нуля','PHP-разработчик с 0 до PRO. Часть 1','SMM-маркетолог от А до Я','UX-дизайн','Анимация интерфейсов','Веб-вёрстка для начинающих 2.0','Веб-дизайн PRO 2.0','Веб-дизайн Базовый','Веб-дизайн с нуля 2.0','Веб-разработчик','Интернет-маркетолог от Ingate',
'Руководитель digital-проектов']
# так как у некоторых студентов нет даты окончания домашки, будем считать, что они пока что ее делают и будем учитывать их в прогрессе поэтому установим в колонке finish_date
# крайнюю дату анализа 2019-07-01
progresss['finish_date'] = progresss['finish_date'].fillna(progresss['finish_date'].max())
# считаем разницу в днях между началом и сдачей домашки. Затем переведем полученное значение в количество месяцев
progresss['duration'] = ((progresss['finish_date'] - progresss['start_date']) / \
                                 np.timedelta64(1, 'M')).apply(np.round) + 1
progresss['month_range'] = progresss.apply(lambda x: pd.date_range(x['start_date'], \
                                                                             periods=x['duration'], \
                                                                             freq="M"), axis=1)
progresss = progresss.explode('month_range')
progresss['month_range'] = progresss['month_range'].dt.strftime('%Y-%m')
progresss.head()
```

```python
df_progress_pivot = progresss.pivot_table('progress_id', index='month_range', \
                                            columns='title', dropna=False, \
                                            aggfunc='count', fill_value=0)
df_progress_pivot.head(30)
```
